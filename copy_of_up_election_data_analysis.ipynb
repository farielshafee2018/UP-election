### #3 How many constituencies are there where the difference between Bharatiya Janata Party and Samajwadi Party is less than 5000

```python
# Filter the dataframe for 'Bharatiya Janata Party' and 'Samajwadi Party'
bjp_sp_votes = bigdat[bigdat['Party'].isin(['Bharatiya Janata Party', 'Samajwadi Party'])]

# Pivot the dataframe to have parties as columns and total votes as values
pivot_votes = bjp_sp_votes.pivot_table(index=['ac_no', 'AC Name', 'District'], columns='Party', values='Total Votes', aggfunc='sum').reset_index()

# Calculate the absolute difference in votes between the two parties
pivot_votes['Vote Difference'] = abs(pivot_votes['Bharatiya Janata Party'] - pivot_votes['Samajwadi Party'])

# Filter for constituencies where the vote difference is less than 5000
close_constituencies = pivot_votes[pivot_votes['Vote Difference'] < 5000]

# Display the result
print("Constituencies where the vote difference between Bharatiya Janata Party and Samajwadi Party is less than 5000:")
display(close_constituencies)

# count total
close_constituencies['ac_no'].count()
```

8. What was the minimum votes difference the Bharatiya Janata party candidate lost with and with which party candidate?

```python
# find all losing BJP members

losing_bjp = bigdat[(bigdat['Party'] == 'Bharatiya Janata Party') & (bigdat['Rank'] == 2)]
losing_bjp.head()

# find the difference of votes between all Bharatiya Janata party members who lost and the candidate who won
# first find the consituency

const = losing_bjp['ac_no'].unique()

# find winners

relev_subset = bigdat[(bigdat['ac_no'].isin(const))]

winner = relev_subset[relev_subset["Rank"]==1]

# merge winner and bjp

#merged_set = pd.merge(losing_bjp, winner, on = "ac_no", how = "inner")
#merged_set.head()

# subset including just those constituencies merging winner and bjp

bjp_loss_subset = pd.merge(losing_bjp, winner[['ac_no', 'Total Votes', 'Candidate', 'Party']], on = "ac_no", suffixes = ("_bjp", "_winner"))

bjp_loss_subset.head()


#find difference and min

diff = abs(bjp_loss_subset['Total Votes_bjp'] - bjp_loss_subset['Total Votes_winner'])
min_diff = diff.min()
print(f'min diff is {min_diff}')



```
